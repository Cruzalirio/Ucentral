```python
from math import comb
import pandas as pd

""" 
1. Escalera Real o Flor Imperial
Se estiman las combinaciones que cumplen la condición:
    1 de los 4 palos disponibles en la baraja
""" 
escalera_real = comb(4,1)

""" 
2. Escalera de Color
Se estiman las combinaciones que cumplen la condición:
    1 de los 4 palos disponibles en la baraja 
    Multiplicado por las diferentes posibilidades que hay de cumplir la condición de escalera de color
    En este caso es importante tener en cuenta que se excluyen 4 cartas entre J y As debido a que seria Escalera Real
""" 
escalera_color = comb(4,1) * comb(9,1)

""" 
3. Póquer
Se estiman las combinaciones que cumplen la condición:
    1 de los 13 valores disponibles en la baraja
    Multiplicado por la selección de las 4 cartas del mismo valor y de los 4 diferentes palos
    Multiplicado por la selección de 1 carta cualquiera entre las 48 restantes
"""
poquer = comb(13,1) * comb(4,4) * comb(48,1)

""" 
4. Full
Se estiman las combinaciones que cumplen la condición:
    1 de los 13 valores disponibles en la baraja
    Multiplicado por la selección de 3 cartas del mismo valor entre los 4 diferentes palos
    Multiplicado por la selección de 1 de los 12 valores restantes disponibles en la baraja
    Multiplicado por la selección de 2 cartas del mismo valor entre los 4 diferentes palos
"""
full = comb(13,1) * comb(4,3) * comb(12,1)  *  comb(4,2)

""" 
5 . Color
Se estiman las combinaciones que cumplen la condición:
    1 de los 4 palos disponibles en la baraja 
    Multiplicado por la selección de 5 cartas de cualquier valor entre las 13 disponibles de 1 palo
    Se le resta la combinatoria de que sea Escalera Real o Escalera de Color
"""
color = comb(4,1) * comb(13,5) - 4 - 36

""" 
6. Escalera
Se estiman las combinaciones que cumplen la condición:
    1 de las 10 cartas con valores mas bajos en la baraja
    Multiplicado 5 veces por la selección de 1 carta entre los 4 palos disponibles
    Se le resta la combinatoria de que sea Escalera Real o Escalera de Color
"""
escalera = comb(10,1) * comb(4,1) * comb(4,1) * comb(4,1) * comb(4,1) * comb(4,1) - 4 - 36

""" 
7. Trío
Se estiman las combinaciones que cumplen la condición:
    1 de los 13 valores disponibles en la baraja
    Multiplicado por la selección de 3 cartas del mismo palo
    Multiplicado por la selección de 2 cartas cualquiera entre las 48 restantes
    Se le resta la combinatoria de que sea Full
"""
trio = comb(13,1) * comb(4,3) * comb(48,2) - 3744

""" 
8. Doble pareja
Se estiman las combinaciones que cumplen la condición:
    2 de los 13 valores disponibles en la baraja
    Multiplicado por la selección de 2 cartas del mismo palo
    Multiplicado por la selección de 2 cartas del mismo palo pero diferente al anterior
    Multiplicado por la selección de 1 valor diferente de los 11 restantes disponibles en la baraja
    Multiplicado por la selección de la carta anterior entre los 4 palos disponibles
"""
doble_pareja = comb(13,2) * comb(4,2) * comb(4,2) * comb(11,1) * comb(4,1)

""" 
9. Pareja
Se estiman las combinaciones que cumplen la condición:
    1 de los 13 valores disponibles en la baraja
    Multiplicado por la selección de 2 cartas del mismo palo
    Multiplicado por la selección de 3 valores diferentes de los 12 restantes disponibles en la baraja
    Multiplicado por la selección de 1 carta del mismo palo
    Multiplicado por la selección de 1 carta del mismo palo pero diferente al anterior
    Multiplicado por la selección de 1 carta del mismo palo pero diferente a las dos anteriores
"""
pareja = comb(13,1) * comb(4,2) * comb(12,3) * comb(4,1) * comb(4,1) * comb(4,1)

""" 
10. Carta Alta
Se estiman las combinaciones que cumplen la condición:
    Se estiman el total de combinacones que contempla 5 cartas entre las 52 disponibles en la baraja
    Se le resta todas las combinaciones posibles anteriormente calculadas
"""
carta_alta = comb(52,5) - 1296420
```

```python
#Diccionario con las jugadas y las combinaciones posibles
poker = {
    "Jugada" : ["Escalera Real", "Escalera Color", "Póquer", "Full", "Color", "Escalera", "Trío", "Doble Pareja", "Pareja", "Carta Alta"],
    "Combinaciones" : [escalera_real, escalera_color, poquer, full, color, escalera, trio, doble_pareja, pareja, carta_alta]
}

#Convertir el diccionario en dataframe
df_poker = pd.DataFrame(data=poker)
df_poker
```
